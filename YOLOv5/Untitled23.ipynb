```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('results.csv')

results.columns = results.columns.str.strip()
print("results.csv:", results.shape)
```

```python
display(results.head())
```

```python
keys = ['box_loss', 'obj_loss', 'cls_loss']

fig, axs = plt.subplots(1, 3, figsize=(15, 2))

plt.suptitle('losses',   size=16, y=1.12)

for k, ax in zip(keys, axs):

  for this_set in ['train', 'val']:

    ax.plot(results['epoch'], results[this_set+'/'+k], label = this_set+'/'+k)

  ax.legend(loc='best')
  ax.grid(True)

plt.subplots_adjust(wspace=0.2)
plt.show()
```

```python
keys = ['metrics/precision', 'metrics/recall', 'metrics/mAP_0.5']
colors = ['r', 'g', 'c']

fig, axs = plt.subplots(1, 3, figsize=(15, 2))

plt.suptitle('losses',   size=16, y=1.12)

for k, ax in zip(range(len(keys)), axs):

  ax.plot(results['epoch'], results[keys[k]], label = keys[k], color = colors[k])

  ax.legend(loc='best')
  ax.grid(True)

plt.subplots_adjust(wspace=0.2)
plt.show()
```
